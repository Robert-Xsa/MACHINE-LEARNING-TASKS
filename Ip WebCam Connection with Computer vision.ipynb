```python
#We are connecting a public IP webcam which is not restricted by any kind of security in accessing its features
#Import Necessary Library
import requests
import cv2
import numpy as np
import imutils

#Class definition
class IP_Web_Camera:
    '''A class used to obtain a streaming video from an IP webCam of Android phone'''
    print("Internet connection must be on!!!!")
    def __init__(self,url = " "):
        self.url = url

    def Capture(self):
        self.url = input("Enter url of your IP webcam: ")
        #An example of url must be in this form
        #An example "http://0.0.0.0:8080/shot.jpg"
        #Where shot.jpg is an extension of your shots taken by camera.
        
        #An Infinity loop to obtain a streaming video
        while True:
            image_captured = requests.get(self.url)
            image_arr = np.array(bytearray(image_captured.content), dtype = np.uint8)
            image = cv2.imdecode(image_arr, cv2.IMREAD_COLOR)
            image = imutils.resize(image, width = 1000, height = 1800)
            cv2.imshow("Android_cam", image)

            #press escape key to exit
            if cv2.waitKey(1) == 27:
                break
                
#Instance of the class
camera = IP_Web_Camera()
camera.Capture()







```
